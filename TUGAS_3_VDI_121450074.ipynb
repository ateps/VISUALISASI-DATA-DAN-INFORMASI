**1. Tentukan Data yang akan Digunakan dalam Visualisasi** \
Saya menggunakan **Tips Dataset** yang sudah tersedia di seaborn. Data ini mencakup informasi tentang total tagihan, tip yang diberikan, jenis kelamin pelanggan, hari makan, waktu makan (lunch/dinner), dan ukuran grup.

**2. Tentukan Tujuan, User, Fungsi, dan Tone**
* **Tujuan**: Menganalisis faktor-faktor yang mempengaruhi besarnya tip yang diberikan pelanggan.
* **User**: Pengelola restoran atau analis bisnis yang tertarik memahami faktor-faktor penentu besarnya tip di restoran.
* **Fungsi**: Membantu restoran atau bisnis dalam menentukan strategi layanan yang dapat meningkatkan besarnya tip yang diberikan.
* **Tone**: Informal namun tetap profesional, dengan visualisasi yang sederhana dan mudah dipahami.

**3. Buat Visualisasi dengan Menggunakan Library Python**

```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the tips dataset
tips = sns.load_dataset("tips")

# Membuat subplots
fig, axes = plt.subplots(3, 2, figsize=(16, 18))

# 1. Boxplot untuk melihat distribusi total bill berdasarkan hari
sns.boxplot(x='day', y='total_bill', data=tips, ax=axes[0, 0])
axes[0, 0].set_title('Distribusi Total Bill Berdasarkan Hari')

# 2. Scatter plot untuk melihat hubungan antara total bill dan tip
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips, ax=axes[0, 1])
axes[0, 1].set_title('Hubungan Total Bill dan Tip Berdasarkan Jenis Kelamin')

# 3. Bar plot untuk melihat rata-rata tip berdasarkan waktu makan
sns.barplot(x='time', y='tip', data=tips, ax=axes[1, 0])
axes[1, 0].set_title('Rata-rata Tip Berdasarkan Waktu Makan (Lunch/Dinner)')

# 4. Heatmap untuk melihat korelasi antar variabel numerik
tips_numeric = tips.select_dtypes(include=['float64', 'int64'])
sns.heatmap(tips_numeric.corr(), annot=True, cmap="coolwarm", linewidths=.5, ax=axes[1, 1])
axes[1, 1].set_title('Korelasi Antar Variabel dalam Dataset Tips')

# 5. Histogram untuk melihat distribusi total bill
sns.histplot(tips['total_bill'], kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Distribusi Total Bill')

# 6. Violin plot untuk melihat distribusi tip berdasarkan jenis kelamin
sns.violinplot(x='sex', y='tip', data=tips, ax=axes[2, 1])
axes[2, 1].set_title('Distribusi Tip Berdasarkan Jenis Kelamin')

# Menyesuaikan layout agar tidak tumpang tindih
plt.tight_layout()
plt.show()
```

**Deskripsi Visualisasi**
1. **Boxplot: Distribusi Total Bill Berdasarkan Hari**
   - **Fungsi**: Menampilkan distribusi jumlah tagihan (total bill) berdasarkan hari. Boxplot ini memperlihatkan median, kuartil atas dan bawah, serta outliers untuk setiap hari.
   - **Interpretasi**: Dari visualisasi ini, kita bisa melihat perbedaan total tagihan di hari-hari yang berbeda dan apakah ada hari tertentu di mana total bill lebih tinggi atau lebih rendah.

2. **Scatter Plot: Hubungan Total Bill dan Tip Berdasarkan Jenis Kelamin**
   - **Fungsi**: Scatter plot ini menunjukkan hubungan antara jumlah tagihan (total bill) dan jumlah tip yang diberikan, dengan perbedaan warna berdasarkan jenis kelamin.
   - **Interpretasi**: Visualisasi ini membantu melihat apakah ada pola yang menunjukkan hubungan antara total bill yang lebih tinggi dengan tip yang lebih besar, serta apakah ada perbedaan berdasarkan jenis kelamin.

3. **Bar Plot: Rata-rata Tip Berdasarkan Waktu Makan (Lunch/Dinner)**
   - **Fungsi**: Bar plot ini menunjukkan rata-rata jumlah tip yang diberikan berdasarkan waktu makan (makan siang atau makan malam).
   - **Interpretasi**: Kita bisa melihat apakah pelanggan memberikan tip yang lebih besar selama makan siang atau makan malam. Ini bisa memberikan insight mengenai kebiasaan memberi tip di waktu yang berbeda.

4. **Heatmap: Korelasi Antar Variabel Numerik**
   - **Fungsi**: Heatmap ini menunjukkan korelasi antar variabel numerik dalam dataset, seperti `total_bill`, `tip`, dan `size` (ukuran grup).
   - **Interpretasi**: Dari heatmap, kita bisa melihat hubungan antara variabel-variabel tersebut. Misalnya, apakah total bill berkorelasi kuat dengan tip yang diberikan, atau bagaimana korelasi antara variabel lainnya.

5. **Histogram: Distribusi Total Bill**
   - **Fungsi**: Histogram ini menunjukkan distribusi jumlah tagihan (total bill). Garis *kde* (Kernel Density Estimation) ditambahkan untuk melihat kepadatan distribusi.
   - **Interpretasi**: Visualisasi ini membantu memahami bagaimana total bill terdistribusi, misalnya apakah kebanyakan tagihan berada di kisaran tertentu atau apakah ada variasi yang besar.

6. **Violin Plot: Distribusi Tip Berdasarkan Jenis Kelamin**
   - **Fungsi**: Violin plot ini menunjukkan distribusi tip berdasarkan jenis kelamin, sekaligus memberikan informasi tentang kepadatan data di berbagai tingkat tip.
   - **Interpretasi**: Violin plot ini membantu membandingkan distribusi tip yang diberikan oleh laki-laki dan perempuan, serta apakah ada perbedaan pola antara dua kelompok tersebut.
